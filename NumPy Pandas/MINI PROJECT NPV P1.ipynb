#### 7. Filter the ‘Emp_Len’ column to extract the numerical value from the string. Hint - Emp_len : < 1year, 2 years , 3 years as 1 , 2, 3 so on.

```python
df['emp_length'].unique()
```

```python
def emp_l(x):
    if (x == '< 1 year')or(x ==  '1 year'):
        return 1
    elif x == '2 years':
        return 2
    elif x == '3 years':
        return 3
    elif x == '4 years':
        return 4
    elif x == '5 years':
        return 5
    elif x == '6 years':
        return 6
    elif x == '7 years':
        return 7
    elif x == '8 years':
        return 8
    elif x == '9 years':
        return 9
    elif x == '10+ years':
        return 10
    else:
        return None
```

```python
df['emp_length']
```

```python
df['emp_length_int'] = df['emp_length'].apply(emp_l)
```

```python
df[['emp_length','emp_length_int']]
```
